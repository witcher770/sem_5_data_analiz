```python
# Импорт необходимых библиотек
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

### Часть 1

```python
df = pd.read_excel('C:\\Users\\Максим\\Desktop\\учеба вуз\\АД\\regression.xlsx', sheet_name='var4') # чтение таблицы
```

#### Линейная регрессия для x1 и y

```python
# --- МОДЕЛЬ 1: Простая линейная регрессия для x1 и y ---

# Разделение данных на признаки (X) и целевую переменную (y) для x1
X1 = df[['x1']]  # Признак x1 в виде DataFrame (требуется для sklearn)
y1 = df['y']     # Целевая переменная y

# Разделение на обучающую и валидационную выборки (75% train, 25% test)
X1_train, X1_valid, y1_train, y1_valid = train_test_split(X1, y1, test_size=0.25, random_state=42)

# Создание и обучение модели линейной регрессии
model1 = LinearRegression()  # Инициализация модели
model1.fit(X1_train, y1_train)  # Обучение модели на обучающих данных

# Предсказание на валидационной выборке
y1_pred = model1.predict(X1_valid)  # Прогноз значений для x1_valid
```

```python
# Создание датафрейма для сравнения истинных и предсказанных значений
df1_results = pd.DataFrame({'Actual': y1_valid, 'Predicted': y1_pred}).reset_index(drop=True)
print("Результаты для модели (x1, y):")
df1_results
```

Чтобы построить линию регрессии необходимо получить коэффициенты с помощью
coef_ и intercept_ (свободный коэффициент), по которым график строится. 

**MSE (Mean Squared Error)** – среднее значение квадратов
ошибок. MSE позволяет оценить, насколько сильно модель отклоняется от истинных значений
целевой переменной. Чем меньше значение MSE, тем более точно модель
предсказывает значения. Однако, MSE имеет особенность оптимизировать ошибки в
среднем, что может привести к недопониманию экстремальных значений или
выбросов.

**средняя абсолютная ошибка MAE (Mean absolute error)** – среднее абсолютное
значение

**среднеквадратичная ошибка RMSE** – квадратный корень из среднего квадрата
ошибок (MSE)

**Коэффициент детерминации R2** – метрика, которую нужно максимизировать,
определяет долю изменений, обусловленных влиянием признаков на целевую
переменную
Метрика коэффициент детерминации хороша тем, что она нормализована, то есть не
превышает единицу - удобно сравнивать разные модели. Например,
метрика MSE может принимать ничем не ограниченные значения больше нуля - это не
всегда удобно.

```python
# Расчет метрик качества регрессии
mse1 = metrics.mean_squared_error(y1_valid, y1_pred)
mae1 = metrics.mean_absolute_error(y1_valid, y1_pred)
rmse1 = np.sqrt(mse1)
r2_1 = metrics.r2_score(y1_valid, y1_pred)

# Вывод метрик
print("\nМетрики качества для модели (x1, y):")
print(f"Mean Absolute Error (MAE): {mae1:.4f}")
print(f"Mean Squared Error (MSE): {mse1:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse1:.4f}")
print(f"R2 Score: {r2_1:.4f}")

# Вывод коэффициентов линейной регрессии (a и b)
print(f"\nКоэффициенты модели (x1, y):")
print(f"a (коэффициент): {model1.coef_[0]:.4f}")
print(f"b (intercept): {model1.intercept_:.4f}")
```

#### Линейная регрессия для x2 и y

```python
# --- МОДЕЛЬ 2: Простая линейная регрессия для x2 и y ---

# Разделение данных на признаки (X) и целевую переменную (y) для x2
X2 = df[['x2']]  # Признак x2 в виде DataFrame
y2 = df['y']     # Целевая переменная y (та же самая)

# Разделение на обучающую и валидационную выборки
X2_train, X2_valid, y2_train, y2_valid = train_test_split(X2, y2, test_size=0.25, random_state=42)

# Создание и обучение модели линейной регрессии
model2 = LinearRegression()  # Инициализация модели
model2.fit(X2_train, y2_train)  # Обучение модели

# Предсказание на валидационной выборке
y2_pred = model2.predict(X2_valid)  # Прогноз значений для x2_valid

# Создание датафрейма для сравнения истинных и предсказанных значений
df2_results = pd.DataFrame({'Actual': y2_valid, 'Predicted': y2_pred}).reset_index(drop=True)
print("Результаты для модели (x2, y):")
display(df2_results.head())

# Расчет метрик качества регрессии
mse2 = metrics.mean_squared_error(y2_valid, y2_pred)
mae2 = metrics.mean_absolute_error(y2_valid, y2_pred)
rmse2 = np.sqrt(mse2)
r2_2 = metrics.r2_score(y2_valid, y2_pred)

# Вывод метрик
print("\nМетрики качества для модели (x2, y):")
print(f"Mean Absolute Error (MAE): {mae2:.4f}")
print(f"Mean Squared Error (MSE): {mse2:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse2:.4f}")
print(f"R2 Score: {r2_2:.4f}")

# Вывод коэффициентов линейной регрессии (a и b)
print(f"\nКоэффициенты модели (x2, y):")
print(f"a (коэффициент): {model2.coef_[0]:.4f}")
print(f"b (intercept): {model2.intercept_:.4f}")
```

**Сравнение моделей по метрикам качества**

Модель (x1, y): MAE: 0.4376 MSE: 0.2982 RMSE: 0.5461 R²: -0.0293

Модель (x2, y): MAE: 0.3789 MSE: 0.2253 RMSE: 0.4747 R²: 0.2221

Анализ сравнения:

По абсолютным ошибкам (MAE): Модель с x2 показывает меньшую среднюю абсолютную ошибку (0.3789 против 0.4376), что означает в среднем ее предсказания ближе к истинным значениям.

По квадратичным ошибкам (MSE/RMSE): Метрики MSE и RMSE также значительно ниже у модели с x2. RMSE = 0.4747 против 0.5461 указывает на то, что типичное отклонение предсказания от истинного значения для модели (x2, y) примерно на 0.07 единицы меньше.

По коэффициенту детерминации (R²): Это ключевой показатель. R² модели (x1, y) равен -0.0293, что меньше нуля. Такой результат означает, что простая линейная модель с x1 работает хуже, чем простое предсказание средним значением целевой переменной y. Модель не объясняет никакой дисперсии в данных. В то же время, модель (x2, y) имеет R² = 0.2221. Это говорит о том, что признак x2 объясняет около 22.21% дисперсии целевой переменной y. Это низкий, но положительный и статистически значимый результат.

**Вывод:** Модель, построенная на признаке x2, является существенно более качественной, чем модель на признаке x1, по всем основным метрикам. Признак x1 практически не имеет линейной связи с целевой переменной y.

**Интерпретация коэффициентов линейной регрессии**

Модель (x1, y):

Уравнение: y = -0.0761 * x1 - 0.5250

Коэффициент a = -0.0761: Очень маленькое по модулю значение, близкое к нулю. Это подтверждает вывод, основанный на метриках: изменение переменной x1 практически не оказывает влияния на целевую переменную y. Отрицательный знак указывает на очень слабую обратную связь.

Константа b = -0.5250: Это прогнозируемое значение y, когда x1 = 0. Фактически, из-за незначимости x1, модель всегда предсказывает значение, близкое к этому числу (что и видно в таблице результатов, где предсказания для разных x1 лежат в узком диапазоне от -0.58 до -0.48).

Модель (x2, y):

Уравнение: y = -0.8549 * x2 - 0.0226

Коэффициент a = -0.8549: Значимый по модулю коэффициент. Он указывает на сильную обратную линейную связь между x2 и y. При увеличении признака x2 на 1 единицу, прогнозируемое значение y уменьшается в среднем на 0.855 единицы.

Константа b = -0.0226: Близка к нулю, что означает, что при x2 = 0, прогнозируемое значение y также будет около нуля. Это логично, если данные центрированы.

**Вывод:** Коэффициенты регрессии наглядно демонстрируют, что x2 является информативным признаком для прогнозирования y, в то время как x1 — нет.

#### Визуализация моделей

**Методы и параметры:**

`plt.figure(figsize=(12, 5))` - создает область для графиков размером 12x5 дюймов

`plt.subplot(1, 2, 1)` - создает сетку 1x2, активирует первый график
- `1, 2, 1` - 1 строка, 2 столбца, позиция 1

`plt.scatter()` - точечный график исходных данных
- `df['x1']` - значения по оси X
- `df['y']` - значения по оси Y  
- `color='blue'` - цвет точек
- `label='Данные'` - метка для легенды

`np.linspace()` - создает равномерно spaced значения
- `df['x1'].min()` - начальное значение
- `df['x1'].max()` - конечное значение  
- `100` - количество точек
- `.reshape(-1, 1)` - преобразует в 2D массив для модели

`model1.predict()` - предсказывает значения y для линии регрессии

`plt.plot()` - строит линию регрессии
- `linewidth=2` - толщина линии

`plt.tight_layout()` - автоматически adjusts отступы между графиками

```python
# --- ВИЗУАЛИЗАЦИЯ ДЛЯ МОДЕЛИ (x1, y) ---

# Точечный график и линия регрессии для x1
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)  # Создание подграфика 1x2, первый график
plt.scatter(df['x1'], df['y'], color='blue', label='Данные')  # Исходные данные
# Построение линии регрессии: используем минимальное и максимальное значение x1 для построения линии
x1_line = np.linspace(df['x1'].min(), df['x1'].max(), 100).reshape(-1, 1)
y1_line = model1.predict(x1_line)
plt.plot(x1_line, y1_line, color='red', linewidth=2, label='Линия регрессии')
plt.xlabel('x1')
plt.ylabel('y')
plt.title('Регрессия: x1 и y')
plt.legend()
plt.grid(True)

# --- ВИЗУАЛИЗАЦИЯ ДЛЯ МОДЕЛИ (x2, y) ---

plt.subplot(1, 2, 2)  # Второй график
plt.scatter(df['x2'], df['y'], color='green', label='Данные')
x2_line = np.linspace(df['x2'].min(), df['x2'].max(), 100).reshape(-1, 1)
y2_line = model2.predict(x2_line)
plt.plot(x2_line, y2_line, color='red', linewidth=2, label='Линия регрессии')
plt.xlabel('x2')
plt.ylabel('y')
plt.title('Регрессия: x2 и y')
plt.legend()
plt.grid(True)

plt.tight_layout()  # Автоматическая настройка расстояний между графиками
plt.show()
```

График "Регрессия: x1 и y": Облако точек не имеет четко выраженного линейного тренда. Линия регрессии является почти горизонтальной (из-за коэффициента a ≈ -0.076), что визуально подтверждает отсутствие линейной зависимости. Точки сильно разбросаны вокруг линии.

График "Регрессия: x2 и y": На этом графике наблюдается видимый нисходящий тренд. Точки в большей степени концентрируются вокруг наклонной линии регрессии, что соответствует значимому коэффициенту a = -0.8549 и лучшим метрикам модели.

#### График предсказанных против истинных значений

**Методы и параметры:**

`plt.scatter(y1_valid, y1_pred)` - точечный график сравнения
- `alpha=0.7` - прозрачность точек

`min()` и `max()` - находят минимальное/максимальное значение для построения линии

`plt.plot([min_val, max_val], [min_val, max_val])` - рисует диагональную линию идеальных предсказаний
- `[min_val, max_val]` - координаты X (от минимума до максимума)
- `[min_val, max_val]` - координаты Y (такие же как X - линия y=x)
- `color='red'` - цвет линии
- `linestyle='--'` - пунктирный стиль
- `label='Идеальная линия'` - метка для легенды

**Интерпретация:**
- Точки на красной линии - идеальные предсказания
- Чем ближе точки к линии, тем точнее модель

**Объяснение про минимум и максимум:**

1. **`min_val = min(y1_valid.min(), y1_pred.min())`**
   - `y1_valid.min()` - минимальное значение из реальных данных
   - `y1_pred.min()` - минимальное значение из предсказаний
   - `min()` - выбирает наименьшее из двух минимумов
   - **Результат:** самая маленькая координата для начала линии

2. **`max_val = max(y1_valid.max(), y1_pred.max())`**
   - `y1_valid.max()` - максимальное значение из реальных данных  
   - `y1_pred.max()` - максимальное значение из предсказаний
   - `max()` - выбирает наибольшее из двух максимумов
   - **Результат:** самая большая координата для конца линии

3. **`plt.plot([min_val, max_val], [min_val, max_val])`**
   - **X координаты:** `[min_val, max_val]` - от минимума до максимума
   - **Y координаты:** `[min_val, max_val]` - такие же значения как X
   - **Уравнение:** y = x (для любой точки: Y = X)

```python
plt.figure(figsize=(12, 5))

# Для модели (x1, y)
plt.subplot(1, 2, 1)
plt.scatter(y1_valid, y1_pred, color='blue', alpha=0.7)  # Синие точки - предсказания

# Красная линия - идеальные предсказания (y = x)
min_val = min(y1_valid.min(), y1_pred.min())
max_val = max(y1_valid.max(), y1_pred.max())
plt.plot([min_val, max_val], [min_val, max_val], color='red', linestyle='--', linewidth=2, label='Идеальная линия')
plt.xlabel('Истинные значения (Actual)')
plt.ylabel('Предсказанные значения (Predicted)')
plt.title('Предсказанные vs Истинные: x1 и y')
plt.legend()
plt.grid(True)

# Для модели (x2, y)
plt.subplot(1, 2, 2)
plt.scatter(y2_valid, y2_pred, color='green', alpha=0.7)
min_val = min(y2_valid.min(), y2_pred.min())
max_val = max(y2_valid.max(), y2_pred.max())
plt.plot([min_val, max_val], [min_val, max_val], color='red', linestyle='--', linewidth=2, label='Идеальная линия')
plt.xlabel('Истинные значения (Actual)')
plt.ylabel('Предсказанные значения (Predicted)')
plt.title('Предсказанные vs Истинные: x2 и y')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

**Графики "Предсказанные vs Истинные":**

Для x1: Точки (синие) расположены практически горизонтально вокруг значения Predicted ≈ -0.5, сильно отклоняясь от красной линии тождества (идеальная линия). Это наглядная иллюстрация плохого качества прогноза.

Для x2: Точки будут в гораздо большей степени тяготеть к красной линии тождества. Хотя идеального совпадения нет, явная положительная корреляция между Actual и Predicted наблюдается, что соответствует R² > 0.

**Вывод:** Визуальный анализ полностью подтверждает количественные метрики. Модель (x2, y) демонстрирует адекватное, хотя и не идеальное, соответствие данным, в то время как модель (x1, y) является совершенно нерелевантной для этих данных.

#### Выводы

Проведенный регрессионный анализ для двух независимых переменных x1 и x2 показал кардинально разные результаты.

Переменная x1 не имеет практически никакой линейной связи с целевой переменной y. Построенная на ее основе линейная регрессионная модель оказалась неэффективной и неинформативной. Ее прогнозы ненамного лучше, чем простое угадывание среднего значения y, о чем свидетельствует коэффициент детерминации R² < 0.

Переменная x2 демонстрирует заметную отрицательную линейную зависимость с переменной y. Модель, построенная на x2, объясняет около 22% дисперсии y и дает существенно более точные прогнозы. Все метрики ошибок (MAE, MSE, RMSE) для этой модели значительно ниже.

Для дальнейшего анализа и построения прогнозных моделей в рамках данной лабораторной работы переменная x2 является перспективным признаком.

y_train, y_valid - целевые переменные

feature_name - имя признака для подписей



```python
# Создадим функцию для удобного обучения и оценки полиномиальной модели
def evaluate_polynomial_model(degree, X_train, X_valid, y_train, y_valid, feature_name):
    """
    Обучает модель полиномиальной регрессии и возвращает метрики и визуализацию
    """
    # Создание pipeline: сначала преобразование в полиномиальные признаки, затем линейная регрессия
    model = Pipeline([
        ('poly', PolynomialFeatures(degree=degree)), # Создает полиномиальные признаки
        ('linear', LinearRegression()) # Обучает линейную регрессию
    ])
    # Обучение модели
    model.fit(X_train, y_train)    
    # Предсказание на валидационной выборке
    y_pred = model.predict(X_valid)
    
    # Расчет метрик
    mae = mean_absolute_error(y_valid, y_pred)
    r2 = r2_score(y_valid, y_pred)
    
    # Визуализация
    plt.figure(figsize=(10, 6))
    
    # Сортируем данные для гладкого построения кривой
    X_sorted = np.sort(X_valid.values.flatten())
    y_pred_sorted = model.predict(X_sorted.reshape(-1, 1))
    
    # Точечный график исходных данных
    plt.scatter(X_valid, y_valid, alpha=0.7, label='Валидационные данные', color='blue')
    
    # Линия полиномиальной регрессии
    plt.plot(X_sorted, y_pred_sorted, color='red', linewidth=2, 
             label=f'Полиномиальная регрессия (степень {degree})')
    
    plt.xlabel(feature_name)
    plt.ylabel('y')
    plt.title(f'Полиномиальная регрессия: {feature_name} и y (степень {degree})')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
    
    return mae, r2, y_pred, model
```

```python
# Подготовка данных для x2
X2 = df[['x2']]
y2 = df['y']

# Разделение на обучающую и валидационную выборки
X2_train, X2_valid, y2_train, y2_valid = train_test_split(X2, y2, test_size=0.25, random_state=42)
```

```python
# Тестирование разных степеней полинома для x2
degrees = [2, 3, 4, 5]
results_x2 = {} # словарь для хранения результатов всех моделей

print("Сравнение различных степеней полинома для (x2, y):")
print("-" * 50)

for degree in degrees:
    print(f"\nСтепень полинома: {degree}")
    mae, r2, y_pred, model = evaluate_polynomial_model(
        degree, X2_train, X2_valid, y2_train, y2_valid, 'x2'
    )
    
    results_x2[degree] = {
        'MAE': mae,
        'R2': r2,
        'model': model,
        'predictions': y_pred
    }
    
    print(f"MAE: {mae:.4f}")
    print(f"R2: {r2:.4f}")
```

##### Функция для оценки моделей

```python
def evaluate_model(model, X_train, X_test, y_train, y_test, model_name):
    """
    Обучает и оценивает модель, возвращает метрики и предсказания
    """
    # Обучение модели
    model.fit(X_train, y_train)
    
    # Предсказания
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    # Метрики для тестовой выборки
    mae = mean_absolute_error(y_test, y_test_pred)
    mse = mean_squared_error(y_test, y_test_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_test_pred)
    
    # Метрики для обучающей выборки (для анализа переобучения)
    r2_train = r2_score(y_train, y_train_pred)
    
    # Создание датафрейма с результатами
    results_df = pd.DataFrame({
        'Actual': y_test,
        'Predicted': y_test_pred,
        'Difference': y_test - y_test_pred
    })
    
    metrics = {
        'MAE': mae,
        'MSE': mse,
        'RMSE': rmse,
        'R2': r2,
        'R2_train': r2_train
    }
    
    return metrics, results_df, y_test_pred, model
```
